```python
from sentence_transformers import SentenceTransformer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentences = [
    # Cricket (4)
    "The batsman scored a century in the final match",
    "The bowler took three wickets in one over",
    "The team won the match by chasing a big target",
    "The captain set an aggressive field placement",

    # Cooking (3)
    "Add spices and cook the vegetables on medium heat",
    "The recipe requires fresh ingredients and herbs",
    "Bake the cake at 180 degrees for 30 minutes",

    # Cybersecurity (3)
    "Use strong passwords to protect your accounts",
    "The system was attacked by a malware virus",
    "Enable two factor authentication for better security"
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=True, cmap='coolwarm')
plt.title("Cosine Similarity Matrix")
plt.show()
```

```python
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 similar sentences
top_indices = similarities.argsort()[-3:][::-1]  # top 3 (includes itself)

print("Top similar sentences:\n")
for idx in top_indices:
    print(f"Sentence: {sentences[idx]}")
    print(f"Similarity Score: {similarities[idx]:.4f}\n")
```
